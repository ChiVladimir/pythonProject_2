```python
import nltk
from textblob import TextBlob

#    nltk
nltk.download('punkt')

#     1
text = "I love programming. It is very exciting and rewarding."

#   TextBlob
blob = TextBlob(text)

#   
sentiment = blob.sentiment

#  
print(f"Text: {text}")
print(f"Polarity: {sentiment.polarity}", f"Subjectivity: {sentiment.subjectivity}", sep='\n')
print("---------------------------------")
#     2
text = "I don't like programming so much. It's very confusing and exhausting."

#   TextBlob
blob = TextBlob(text)

#   
sentiment = blob.sentiment

#  
print(f"Text: {text}")
print(f"Polarity: {sentiment.polarity}", f"Subjectivity: {sentiment.subjectivity}", sep='\n')
print("---------------------------------")

#     3
text = "We love programming so much. It's very interesting and useful."

#   TextBlob
blob = TextBlob(text)

#   
sentiment = blob.sentiment

#  
print(f"Text: {text}")
print(f"Polarity: {sentiment.polarity}", f"Subjectivity: {sentiment.subjectivity}", sep='\n')
print("---------------------------------")
```
